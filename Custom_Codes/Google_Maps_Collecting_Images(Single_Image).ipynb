```python
import requests
import time
import json

api_key = "AIzaSyBnuSYPHgNdunNlVVKwhWqMaRzv-Ox7k7k"

# # Coordinates for the image
# latitude = 46.061749185361286
# longitude = -90.43341155638159
# center = f"{latitude},{longitude}"


with open("../Adnan_Tasks/addresses.json") as f:
    addresses = json.load(f)


zoom = 18
scale = 2
size = "1280x1280"
map_type = "satellite"


for address in addresses:
    print(address)
    center = address.replace(" ", "+")
    url = (
        f"https://maps.googleapis.com/maps/api/staticmap?"
        f"center={center}"
        f"&zoom={zoom}"
        f"&size={size}"
        f"&scale={scale}"
        f"&maptype={map_type}"
        f"&key={api_key}"
    )

    
    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(f"../Adnan_Tasks/Images/Custome_Collected_Adnan_PDF/{address}.png", "wb") as f:
            f.write(response.content)
        print(f"Image saved as {address}.png")
    except Exception as e:
        print(f"Error: {e}")
```
